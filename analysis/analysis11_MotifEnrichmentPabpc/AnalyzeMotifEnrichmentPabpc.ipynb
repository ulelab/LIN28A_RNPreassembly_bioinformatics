```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from glob import glob
import sys
from sklearn.preprocessing import StandardScaler

scriptpath = "../../src/"
sys.path.append(os.path.abspath(scriptpath))
import KmerGroups as kg
import PekaLib as peka
```

```python
# Import PEKA results for PABPC data
PekaFiles = sorted(glob('../../data/Pabpc1Pabpc4Iclip_2022/PEKA/*.tsv')) + sorted(glob('../../results/20221006_Pabpc1Peka/PekaOutput/*DKO_*5mer_distribution_UTR3.tsv'))

# Save to
outpath = '../../results/AnalyzeMotifEnrichmentForPabpc'
os.makedirs(outpath, exist_ok=True)
srcOut = '../../results/SourceData/'
os.makedirs(srcOut, exist_ok=True)
```

```python
PekaFiles
```

```python
DfScores = peka.GetPEKAScoresFromTsv(PekaFiles).fillna(0)
```

```python
scaler = StandardScaler()
scaler.fit(DfScores)
DfZScores = pd.DataFrame(scaler.transform(DfScores), columns = DfScores.columns, index = DfScores.index)
```

```python
DfZScores.head()
```

```python
SignificantKmersDict = peka.GetSignificantKmersFileList(PekaFiles)
```

```python
SignificantKmersDict
```

```python
AllSigKmers = []
for kmers in SignificantKmersDict.values():
    AllSigKmers.extend([ k for k in kmers if k not in AllSigKmers])
AllSigKmers = sorted(set(AllSigKmers))
```

```python
len(AllSigKmers)
```

# Compare significant k-mers between KO and DOX for PABPC1 and PABPC4
Combine all significant k-mers and cluster them into groups. Then compare z-cores in each group of k-mers across samples


```python
DfSigKmers = DfZScores.loc[AllSigKmers]
DfSigKmers.columns = [x.split('_5mer_dist')[0] for x in DfSigKmers.columns]
DfSigKmersMelt = DfSigKmers.melt(var_name='File', value_name='z-score')
DfSigKmersMelt['Sample'] = DfSigKmersMelt.File.apply(lambda x: x.split('_5mer_distribution')[0])
```

```python
orderCond = ['DOX', 'KO']
orderProt = ['C1', 'C4_Proteintech', 'C4_Benthyl']
colOrd = []
for p in orderProt:
    cols = sorted([x for x in DfSigKmers.columns if p in x])
    colOrd.extend(cols)
DfSigKmers = DfSigKmers[colOrd]
```

```python
DfSigKmers.head()
```

```python
fig = sns.clustermap(DfSigKmers, col_cluster=False)
fig.savefig(f'{outpath}/ZScoreHeatmap_NaNToZero.pdf', bbox_inches='tight')
```

## Count A-rich and U-rich significant k-mers in each sample and plot counts

```python
rows = []
for k, vals in SignificantKmersDict.items():
    a_rich = []
    u_rich = []
    for v in vals:
        if list(v).count('A') >= 3:
            a_rich.append(v)
        elif list(v).count('U') >= 3:
            u_rich.append(v)
        else:
            pass
    row = [k, len(a_rich), len(u_rich), len(vals)]
    rows.append(row)
DfCountGroups = pd.DataFrame(rows, columns=['File', 'A-rich', 'U-rich', 'N significant total'])
```

```python
DfCountGroups = DfCountGroups.melt(var_name='Kmer group', value_name='N Significant kmers', id_vars=['File', 'N significant total'])
DfCountGroups['% significant k-mers'] = DfCountGroups['N Significant kmers'].divide(DfCountGroups['N significant total'], axis='index') * 100
DfCountGroups['Sample'] = DfCountGroups['File'].apply(lambda x: x.split('_5mer_')[0])
DfCountGroups['Condition'] = DfCountGroups['File'].apply(lambda x: x.split('_')[0] if x.split('_')[0] in ['KO', 'DOX'] else '.')
DfCountGroups['Experiment'] = DfCountGroups['Sample'].apply(lambda x: x.strip('12_DOXKO'))
```

```python
DfCountGroups.loc[DfCountGroups.Sample == 'PABPC1_ESC_DKO_doxGFPLin28A_grouped_cdna_mm39', 'Condition'] = 'DOX'
DfCountGroups.loc[DfCountGroups.Sample == 'PABPC1_ESC_DKO_doxGFPLin28A_grouped_cdna_mm39', 'Experiment'] = '2iL PABPC1'

DfCountGroups.loc[DfCountGroups.Sample == 'PABPC1_ESC_DKO_nodox_grouped_cdna_mm39', 'Condition'] = 'KO'
DfCountGroups.loc[DfCountGroups.Sample == 'PABPC1_ESC_DKO_nodox_grouped_cdna_mm39', 'Experiment'] = '2iL PABPC1'
```

```python
# Save source data
dfSrc = DfCountGroups.copy()
# Drop C4_Benthyl experiments
dfSrc = dfSrc[dfSrc.Experiment != 'C4_Benthyl']

# Add a biological replicate column - last character of the sample name
dfSrc['Biological Replicate'] = dfSrc['Sample'].apply(lambda x: int(x[-1]))
# Add a protein column
dfSrc['Protein'] = dfSrc.Sample.apply(lambda x: 'PABPC1' if '_C1' in x else 'PABPC4')

# Drop sample and experiment column
dfSrc = dfSrc.drop(columns=['Experiment', 'Sample'])

# Reorder columns
dfSrc = dfSrc[['Protein', 'Condition', 'Biological Replicate', 'File', 'Kmer group', 'N significant total', 'N Significant kmers', '% significant k-mers']]
# Rename columns
dfSrc.columns = ['Protein', 'Condition', 'Biological Replicate', 'File', 'kmer-group', 'N significant all kmers', 'N significant kmer-group', '% significant kmer-group']
# Save source data
dfSrc.to_csv(f'{srcOut}/PABPC_PEKA_A-U_rich_kmer_proportions.csv', index=False)
```

```python
sns.set_style('whitegrid')
fig = sns.catplot(
    data=DfCountGroups, x='Condition', y='N Significant kmers', hue='Kmer group', col='Experiment', kind='bar', col_order=['C1_Crick', '2iL PABPC1', 'C4_Benthyl', 'C4_Proteintech']
)
fig.savefig(f'{outpath}/n_SignificantKmers.pdf', bbox_inches='tight')
```
